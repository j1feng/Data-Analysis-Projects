## Datasets

Cardiovascular Dieases dataset https://www.kaggle.com/sulianova/cardiovascular-disease-dataset
```
id          - ID number  
age         - days
gender      - 1 - women, 2 - men

````python
cvd_dir = "datasets/cvd_dataset/cardio_train.csv"
df = pd.read_csv(cvd_dir, sep=';')
# print("Raw Dataset shape: {}".format(df.shape))
# print("Raw Dataset Features: \n{}".format(df.columns.values))
df
````

# Data Cleaning
Observing the data, most data is clean and straight forward, but age column is a bit weird since the unit is in days. Here we convert it to years to make more sense. In addition `id` column is not really useful so we can drop it.

````python
if "id" in df:
    df = df.drop("id", axis = 1)
df["age_year"] = df["age"].map(lambda age: round(age / 365))
df
````

# Data Normalization

````python
# Splitting and shuffling data

df2 = sk.utils.shuffle(df)

scaler = MinMaxScaler()

# Normalizing fields for regression
df2["gender"]   = df["gender"] - 1
df2 = df2.drop("age", axis = 1)

df_normalized = scaler.fit_transform(df2)
# df2["age_year"] = min_max_scaler.fit_transform(df2["age_year"].values)
# df2["height"]   = min_max_scaler.fit_transform(df2["height"].values)
# df2["weight"]   = min_max_scaler.fit_transform(df2["weight"].values)
# df2["ap_hi"]    = min_max_scaler.fit_transform(df2["ap_hi"].values)
# df2["ap_low"]   = min_max_scaler.fit_transform(df2["ap_low"].values)

df_normalized = pd.DataFrame(df_normalized)
df_normalized.columns = df2.columns
df_normalized
````

## Training & Testing set

````python


'''I (Zain) was thinking we could separate male and female observations
for experimental validity. If so we could uncomment this: '''

# labels_gen1 = df_gen1.cardio
# inputs_gen1 = df_gen1.drop(columns=['id', 'cardio'])

# labels_gen2 = df_gen2.cardio
# inputs_gen2 = df_gen2.drop(columns=['id', 'cardio'])

# data_dict = {
#     'gender1_inputs': inputs_gen1,
#     'gender1_labels': labels_gen1,
#     'gender2_inputs': inputs_gen2,
#     'gender2_labels': labels_gen2
# }

labels = df_normalized["cardio"].to_numpy().transpose()
inputs = df_normalized.drop('cardio', axis = 1)
print("Input shape: {0}, Labels shape: {1}".format(inputs.shape, labels.shape))

# Split into folds for training/testing
frac = 0.8
n = labels.shape[0]
split_idx = int(frac * n)

train_inputs = inputs[:split_idx]
train_labels = labels[:split_idx]
test_inputs = inputs[split_idx:n]
test_labels = labels[split_idx:n]

print("Training shapes:")
print("Train inputs shape: {0}, Train labels shape: {1}".format(train_inputs.shape, train_labels.shape))

print("")

print("Test shapes:")
print("Test inputs shape: {0}, Test labels shape {1}".format(test_inputs.shape, test_labels.shape))
````

### A simple linear regression

````python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(train_inputs, train_labels)

# Make predictions using the testing set
y_pred = regr.predict(test_inputs)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(test_labels, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(test_labels, y_pred))
````

##### Doesn't look good lol
The linear model is no good, probably because the factors are not independent on each other, for example, maybe someone would have a higher rate of getting cardiovascular disease if he/she drinks, smokes, and doesn't do exercise.
